### **Ürün vs Satış Miktarı**

### **5) OnlineRetail2.csv veri setini kullanarak günlere göre (Türkçe gün isimleriyle) ortalama satışları gösteren bar grafiği çiziniz.**

```python
df3 = pd.read_csv("C:\\Users\\Ahmet\\Documents\\Vs_Code\\Python\\OnlineRetail2.csv")
```

```python
df3.head()
```

```python
df3.columns
```

```python
df3.info()
```

```python
df3['InvoiceDate'] = pd.to_datetime(df3['InvoiceDate'])
```

```python
df3['TotalPrice'] = df3['Quantity'] * df3['UnitPrice']
```

```python
df3['day'] = df3.InvoiceDate.dt.day
```

```python
df3.head()
```

```python
# Günlerin isimleri ile rakamları eşleştiren fonksiyon
def day_switch(i):
        switcher={
                
        1: 'Pazartesi',
        2: 'Salı',
        3: 'Çarşamba',
        4: 'Perşembe',
        5: 'Cuma',
        6: 'Cumartesi',
        7: 'Pazar'

             }
        return switcher.get(i)
```

```python
# Pandas dataframe day series içinde dolaşarak paterne uygun eşleştirme yapalım
days = []
for i in df3.day:
    days.append(day_switch(i))
```

```python
# Oluşturduğumuz paterni taşıyan listeyi yeni bir sütun olarak dataframe e ekleyelim
df3["day_names"] = days
```

```python
df3.head()
```

```python
#gün bazlı satışları toplamını alıp gruplandırarak büyükten küçüğe yazar
df3_day_grp = df3 \
.groupby(['day_names']).TotalPrice \
.agg(['sum']) \
.sort_values('sum',ascending=False) 
```

```python
df3_day_grp.head()
```

```python
df3_day_grp.index.unique()
```

```python
x = df3_day_grp.index.unique()
y = df3_day_grp['sum']


y_pos = range(len(x))
performance = y
plt.figure(figsize=(16,8)) 
plt.bar(y_pos, performance, align='center', alpha=0.5) # alfa saydamlık derecesi
plt.xticks(y_pos, x, fontsize=14)
plt.xlabel('Günler', fontsize=18)
plt.xticks(rotation=90)
plt.ylabel('Ortalama Satışlar', fontsize=18)
plt.title('Günlere Göre Toplam Satışlar', fontsize=22)
plt.show()
```
